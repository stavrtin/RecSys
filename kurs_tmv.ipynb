```python
result = data_ret_test.groupby('user_id')['item_id'].unique().reset_index()
result.columns=['user_id', 'actual']
result.head(2)
```

```python
def random_recommendation(items, n=5):
    """Случайные рекомендации"""
    
    items = np.array(items)
    recs = np.random.choice(items, size=n, replace=False)
    
    return recs.tolist()
```

```python
%%time

# Можно так делать, так как рекомендация не зависит от юзера
popular_recs = random_recommendation(data_ret_test, n=5)

result['random_recommendation'] = result['user_id'].map(lambda x: popular_recs)
result.head(2)
```
